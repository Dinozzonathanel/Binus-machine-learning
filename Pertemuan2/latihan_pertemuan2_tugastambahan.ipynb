# Import library dan Dataset



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("penguins")
```

# Eksplorasi Awal Dataset dan Visualisasi data

```python
df.head()
df.info()
df.describe()
```

# Note, harus diperhatikan
> Histogram (body_mass_g)

> Boxplot (body_mass_g)

> Heatmap korelasi (body_mass_g)

## Histogram

```python
sns.histplot(df["body_mass_g"].dropna())
plt.show()
```

# Boxplot

```python
sns.boxplot(x=df["body_mass_g"])
plt.show()
```

## Heatmap Korelasi

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```
